```python
import pandas, os, overpass, sys, requests, json, matplotlib.pyplot as plt, numpy as np

pandas.__version__
```

```python
overpass_url = "http://overpass-api.de/api/interpreter"
```

```python
overpass_query = """
[out:json];
area["ISO3166-1"="GB"][admin_level=2];
(node["amenity"="bar"](area);
 way["amenity"="bar"](area);

);
out center;
"""
```

```python
response = requests.get(overpass_url, 
                        params={'data': overpass_query})
```

```python
data = response.json()
```

```python
data
```

```python
# Collect coords into list
coords = []
for element in data['elements']:
  if element['type'] == 'node':
    lon = element['lon']
    lat = element['lat']
    coords.append((lon, lat))
    
  elif 'center' in element:
    lon = element['center']['lon']
    lat = element['center']['lat']
    coords.append((lon, lat))
```

This Cell above assigns the lat and long to coords 

```python
coords
```

```python
X = np.array(coords)
```

```python
plt.plot(X[:, 0], X[:, 1], 'x')
plt.title('Bars in Britain')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.axis('equal')
plt.show()
```
